### 5) Transformasi Data

Pada dataset driver behaviour, kami menggunakan transformasi data yakni normalisasi. Normalisasi berfungsi untuk menormalkan rentang nilai dari setiap fitur pada dataset sehingga setiap fitur tersebut memiliki nilai yang sama. Teknik normalisasi ini diterapkan pada fitur-fitur seperti acceleration dan rotation yang memiliki nilai rentang yang berbeda-beda.

### Normalisasi MinMax

```
def cetak_rentang(df_input):
  for fitur in df_input:
    max = df_input[fitur].max()
    min = df_input[fitur].min()
    print("Rentang fitur ",fitur," adalah ",max-min)
```

```
cetak_rentang(data)
```

```
def minmax(df_input):
  for fitur in df_input:
    max = df_input[fitur].max()
    min = df_input[fitur].min()
    df_input[fitur] = (df_input[fitur]-min)/(max-min)
  return df_input
```

```
data_minmax = minmax(data)
```

```
data_minmax.head()
```

```
cetak_rentang(data_minmax)
```

```
data_minmax.describe()
```

### Normalisasi Z-score


```
def zscore(df_input):
  for fitur in df_input:
    mean = df_input[fitur].mean()
    std = df_input[fitur].std()
    df_input[fitur] = (df_input[fitur]-mean)/std
  return df_input
```

```
data_zscore = zscore(data)
```

```
data_zscore.head()
```

```
data_zscore.describe()
```
